```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

data = data_amusement_cities
# Ensure the 'Year' column is a datetime object
if 'Year' in data.columns:
    data['Year'] = pd.to_datetime(data['Year'], format='%Y')  # Convert 'Year' to datetime
    data.set_index('Year', inplace=True)  # Set 'Year' as the index
else:
    print("'Year' column not found in the dataset. Please check the file.")

# Function to plot time series
def plot_time_series(df, column, city_name):
    """
    Plot the time series of a specified column.
    
    Parameters:
        df (pd.DataFrame): DataFrame containing the data.
        column (str): Column name to plot.
        city_name (str): Name of the city for title.
    """
    plt.figure(figsize=(10, 5))
    plt.plot(df.index, df[column], marker='o', linestyle='-')  # Use df.index directly
    plt.title(f'{column} Over Years - {city_name}')
    plt.xlabel('Year')
    plt.ylabel(column)
    plt.grid(True)
    plt.show()

# Split data by city
df_chennai = data[data['City'] == 4]
df_delhi = data[data['City'] == 2]

# Plot Average Daily Footfalls for Chennai and Delhi
plot_time_series(df_chennai, 'Average Daily Footfalls', 'Chennai')
plot_time_series(df_delhi, 'Average Daily Footfalls', 'Delhi')

# -------------------------------
# Time Series Forecasting with ARIMA
# -------------------------------

def forecast_footfalls_arima(ts, order=(1,1,1)):
    """
    Fit ARIMA model and forecast footfalls.
    
    Parameters:
        ts (pd.Series): Time series data.
        order (tuple): ARIMA order (p, d, q).
        
    Returns:
        float: Forecasted footfalls.
        ARIMAResultsWrapper: Fitted ARIMA model.
    """
    model = ARIMA(ts, order=order)
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=1)
    return forecast.iloc[0], model_fit

# Chennai Footfalls Forecast
chennai_footfall_forecast, chennai_model = forecast_footfalls_arima(df_chennai['Average Daily Footfalls'], order=(1,1,1))

# Delhi Footfalls Forecast
delhi_footfall_forecast, delhi_model = forecast_footfalls_arima(df_delhi['Average Daily Footfalls'], order=(1,1,1))

# -------------------------------
# Calculate Predicted Revenue
# -------------------------------

def calculate_revenue(footfalls, ticket_price):
    """
    Calculate predicted revenue based on footfalls and ticket price.
    
    Parameters:
        footfalls (float): Forecasted average daily footfalls.
        ticket_price (float): Average ticket price in INR.
        
    Returns:
        float: Predicted revenue in NPR Million.
        int: Total visitors for the year.
    """
    total_visitors = int(footfalls * 365)  # Assuming 365 days in a year
    revenue = (ticket_price * total_visitors) / 1e6  # Convert to NPR Million
    return revenue, total_visitors

# Expected Ticket Prices for 2024
chennai_ticket_price = 1268  # Expected Ticket Price for Chennai in 2024
delhi_ticket_price = 960      # Expected Ticket Price for Delhi in 2024

# Calculate Revenue
chennai_revenue, chennai_total_visitors = calculate_revenue(chennai_footfall_forecast, chennai_ticket_price)
delhi_revenue, delhi_total_visitors = calculate_revenue(delhi_footfall_forecast, delhi_ticket_price)

# -------------------------------
# Calculate Growth Rates
# -------------------------------

def calculate_growth_rate(forecast, last_value):
    """
    Calculate the growth rate from the last observed value to the forecasted value.
    
    Parameters:
        forecast (float): Forecasted value.
        last_value (float): Last observed value.
        
    Returns:
        float: Growth rate in percentage.
    """
    growth_rate = ((forecast - last_value) / last_value) * 100
    return growth_rate

# Get last observed footfalls and revenue
last_year_chennai_footfalls = df_chennai['Average Daily Footfalls'].iloc[-1]
last_year_delhi_footfalls = df_delhi['Average Daily Footfalls'].iloc[-1]

# Calculate growth rates for footfalls
chennai_footfall_growth = calculate_growth_rate(chennai_footfall_forecast, last_year_chennai_footfalls)
delhi_footfall_growth = calculate_growth_rate(delhi_footfall_forecast, last_year_delhi_footfalls)

# Calculate last year's revenue
# Since revenue was derived from footfalls, recalculate last year's revenue
last_year_chennai_revenue, _ = calculate_revenue(last_year_chennai_footfalls, chennai_ticket_price)
last_year_delhi_revenue, _ = calculate_revenue(last_year_delhi_footfalls, delhi_ticket_price)

# Calculate growth rates for revenue
chennai_revenue_growth = calculate_growth_rate(chennai_revenue, last_year_chennai_revenue)
delhi_revenue_growth = calculate_growth_rate(delhi_revenue, last_year_delhi_revenue)

# -------------------------------
# Display the Results
# -------------------------------

print("### Predicted Metrics for 2024 ###\n")
print(f"Delhi:")
print(f"  - Forecasted Average Daily Footfalls: {delhi_footfall_forecast:.2f}")
print(f"  - Total Visitors (2024): {delhi_total_visitors}")
print(f"  - Predicted Revenue (2024): ₹{delhi_revenue:.2f} Million")
print(f"  - Expected Footfall Growth Rate: {delhi_footfall_growth:.2f}%")
print(f"  - Expected Revenue Growth Rate: {delhi_revenue_growth:.2f}%\n")

print(f"Chennai:")
print(f"  - Forecasted Average Daily Footfalls: {chennai_footfall_forecast:.2f}")
print(f"  - Total Visitors (2024): {chennai_total_visitors}")
print(f"  - Predicted Revenue (2024): ₹{chennai_revenue:.2f} Million")
print(f"  - Expected Footfall Growth Rate: {chennai_footfall_growth:.2f}%")
print(f"  - Expected Revenue Growth Rate: {chennai_revenue_growth:.2f}%\n")

# -------------------------------
# Visualization of Forecasts
# -------------------------------

def plot_forecast_arima(ts, forecast, model_fit, city_name):
    """
    Plot historical data with ARIMA forecast.
    
    Parameters:
        ts (pd.Series): Time series data.
        forecast (float): Forecasted footfalls.
        model_fit: Fitted ARIMA model.
        city_name (str): Name of the city for title.
    """
    plt.figure(figsize=(10, 5))
    plt.plot(ts.index, ts, label='Historical Footfalls', marker='o')
    
    # Create forecast index
    forecast_year = ts.index.max().year + 1
    plt.plot(forecast_year, forecast, label='Forecasted Footfalls', marker='o', linestyle='--', color='red')
    
    # Confidence intervals
    pred_uc = model_fit.get_forecast(steps=1)
    pred_ci = pred_uc.conf_int()
    plt.fill_between([forecast_year - 0.1, forecast_year + 0.1],
                     pred_ci.iloc[0, 0],
                     pred_ci.iloc[0, 1],
                     color='pink', alpha=0.3)
    
    plt.title(f'Average Daily Footfalls Forecast for {city_name} (ARIMA)')
    plt.xlabel('Year')
    plt.ylabel('Average Daily Footfalls')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot Forecast for Chennai
plot_forecast_arima(df_chennai['Average Daily Footfalls'], chennai_footfall_forecast, chennai_model, 'Chennai')

# Plot Forecast for Delhi
plot_forecast_arima(df_delhi['Average Daily Footfalls'], delhi_footfall_forecast, delhi_model, 'Delhi')
```
